# 1. Provide the trend chart for the number of complaints at monthly and daily granularity levels

Trend chart complaints based on Date of the Month

```python
df.dtypes
```

```python
# Add Date Month year with Time and save it into Date_Full
df["Date_Full"] = df["Date_month_year"] + ' ' +  df["Time"]
```

```python
#Convert Date_Full and Date_month_year to Datetime Format
df["Date_Full"] = pd.to_datetime(df["Date_Full"])
df["Date_month_year"] = pd.to_datetime(df["Date_month_year"])
df_monthly = df.set_index(df["Date_Full"])
```

```python
# Provide the trend chart for the number of complaints at monthly granularity levels.
#Increase Graph Size
plt.figure(figsize=(14,6))
plt.suptitle('Number of complaints at Monthly granularity levels')
plt.ylabel('Number of Complaints')
df_monthly.groupby(pd.Grouper(freq="M")).size().plot(color='red')
```

```python
# Provide the trend chart for the number of complaints at daily granularity levels.
df['Day of Month'] = pd.to_datetime(df['Date'])
df_daily = df.set_index(df["Day of Month"])
#Increase Graph Size
plt.figure(figsize=(14,6))
plt.suptitle('Number of complaints at Daily granularity levels')
plt.ylabel('Number of Complaints')
df_daily.groupby(pd.Grouper(freq="D")).size().plot(color='red')
```

# Task 4: Create a new categorical variable with value as Open and Closed. Open & Pending is to be categorized as Open and Closed & Solved is to be categorized as Closed.

```python
# Checking how many unique values are under Status Column 
df['Status'].unique()
```

```python
# Converting as per Instruction (Task 4) into New Column without changing the main data so that I can use the main data
# in Future
df['New_Status']= ["Open" if Status=="Open" or Status=="Pending" else 
                                      "Closed" for Status in df["Status"]]
```

```python
# Checking whether all status updated or not
df['New_Status'].unique()
```

```python
df_status_by_state = pd.crosstab(df.State,df.New_Status)
```
